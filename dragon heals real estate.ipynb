```python
import numpy as np
np.random.seed(42)
def split_train_test(data,test_ratio):
    shuffeled = np.random.permutation(len(data))
    print(shuffeled)
    test_set_size = int(len(data)*test_ratio)
    test_ind = shuffeled[:test_set_size]
    train_ind = shuffeled[test_set_size:]
    return data.iloc[train_ind],data.iloc[test_ind]
```

```python
train_set,test_set = split_train_test(housing ,0.2)
```
